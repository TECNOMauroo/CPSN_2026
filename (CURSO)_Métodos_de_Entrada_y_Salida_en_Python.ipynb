#1. Método print
El método print en Python se utiliza para mostrar información en la pantalla. Es una de las funciones más básicas y utilizadas en la programación.

```python
#Uso Básico de print - El uso más sencillo de print es mostrar una cadena de texto:
print("Hola, mundo!")
```

```python
# Imprimir Múltiples Elementos
# Puedes usar print para mostrar múltiples elementos separados por comas. Python automáticamente añade un espacio entre ellos:
nombre = "Juan"
edad = 30
print("Nombre:", nombre, "Edad:", edad)
```

```python
# Parámetro sep
"""
El parámetro sep en la función print() de Python define el separador que se coloca entre
múltiples elementos que deseas imprimir. Por defecto, su valor es un espacio en blanco (' '),
pero puedes personalizarlo con cualquier carácter, texto, o incluso dejarlo vacío para unir el texto.
"""
print("Nombre:", nombre, "Edad:", edad, sep="  ")
```

```python
# Parámetro end
# El parámetro end permite definir lo que se imprimirá al final de la salida. Por defecto, print añade una nueva línea
print("Hola", end=" ")
print("mundo!")
```

```python
# Formato de Cadenas
# Para mostrar mensajes más complejos, puedes utilizar el método format o las f-strings (en Python 3.6+)

"""
El método .format() en Python permite insertar variables y valores dentro de una
cadena de texto (string) utilizando llaves {} como marcadores de posición.
Reemplaza el texto dentro de las llaves con los datos pasados como argumentos
"""
print("Nombre: {} Edad: {}".format(nombre, edad))


"""
Las f-strings (cadenas literales formateadas) introducidas en Python 3.6 son la forma más rápida,
legible y eficiente de insertar variables y expresiones dentro de un texto.
Se crean anteponiendo la letra f (o F) antes de las comillas y colocando las variables entre llaves {}.
"""
print(f"Nombre: {nombre} Edad: {edad}")
```

#2. Método pprint
El método pprint (pretty print) del módulo pprint se utiliza para imprimir estructuras de datos complejas de manera más legible. Esto es particularmente útil para listas y diccionarios anidados.

```python
# Primero, necesitas importar el módulo pprint. Luego, puedes utilizar pprint
# para imprimir una estructura de datos compleja

from pprint import pprint
datos = {
    "nombre": "Juan",
    "edad": 30,
    "hobbies": ["leer", "correr", "programar"],
    "educacion": {
        "primaria": "Escuela A",
        "secundaria": "Escuela B",
        "universidad": "Universidad C"
    }
}

pprint(datos)
```

```python
#Parámetros de pprint
#El método pprint permite ajustar la anchura de la salida y el nivel de anidamiento.

"""
Este ajuste limita la anchura de cada línea a 40 caracteres y el nivel de anidamiento
a 2 niveles, haciendo la salida aún más legible en ciertos casos.
"""
pprint(datos, width=40, depth=2)
```

#3. Método input
El método input se utiliza para capturar la entrada del usuario. Esta función espera a que el usuario ingrese datos y luego devuelve lo que se haya escrito como una cadena de texto.

```python
#Uso Básico de input
#Para capturar la entrada del usuario, simplemente llama a input con un mensaje opcional

nombre = input("¿Cuál es tu nombre? ")
print(f"Hola, {nombre}!")
```

```python
#Conversión de Tipos
# La entrada de input siempre es una cadena, por lo que
# si necesitas un número, deberás convertirlo

edad = input("¿Cuántos años tienes? ")
edad = int(edad)  # Convertir a entero
print(f"Tienes {edad} años.")
```

```python
#Si el usuario ingresa algo que no puede convertirse a un número,
#el programa generará un error, por lo que es buena práctica manejar las excepciones:

try:
    edad = int(input("¿Cuántos años tienes? "))
    print(f"Tienes {edad} años.")
except ValueError:
    print("Por favor, ingresa un número válido.")
```

#4. Cadenas y Métodos de Cadenas
Las cadenas en Python son secuencias de caracteres. Python ofrece una amplia variedad de métodos para trabajar con cadenas. Vamos a explorar algunos de los métodos más comunes y útiles.

```python
#Concatenación de Cadenas
#Puedes concatenar (unir) cadenas utilizando el operador +:
saludo = "Hola"
nombre = "Juan"
mensaje = saludo + " " + nombre
print(mensaje)
```

```python
# 1. Upper y lowe
# Estos métodos convierten una cadena a mayúsculas o minúsculas respectivamente.
texto = "Hola Mundo"
print(texto.upper())  # "HOLA MUNDO"
print(texto.lower())  # "hola mundo"
```

```python
# 2. len
# Devuelve la longitud de una cadena.
print(len(texto))  # 10
```

```python
# 3. strip, lstrip y rstrip
# Estos métodos eliminan los espacios en blanco desde ambos extremos,
# el inicio o el final de una cadena.
print(texto.strip())   # "Hola Mundo"
print(texto.lstrip())  # "Hola Mundo   "
print(texto.rstrip())  # "   Hola Mundo"
```

```python
# 4. split
# split divide una cadena en una lista de subcadenas,
# y join une una lista de subcadenas en una sola cadena.
texto = "Hola, Mundo, Python"
lista = texto.split(", ")  # ['Hola', 'Mundo', 'Python']
print(lista)

nuevo_texto = " ".join(lista)
print(nuevo_texto)  # "Hola Mundo Python"
```

```python
# 5. replace
# Reemplaza una subcadena por otra.
texto = "Hola Mundo"
nuevo_texto = texto.replace("Mundo", "Python")
print(nuevo_texto)  # "Hola Python"
```

```python
# 6. find, index
# Buscan una subcadena dentro de una cadena
# y devuelven la posición de su primer carácter.
# find devuelve -1 si no encuentra la subcadena, mientras que index genera una excepción.

texto = "Hola Mundo"
print(texto.find("Mundo"))  # 5
print(texto.index("Mundo"))  # 5
```

```python
# 7. startswith, endswith
# Verifican si una cadena comienza o termina con una subcadena específica
texto = "Hola Mundo"
print(texto.startswith("Hola"))  # True
print(texto.endswith("Mundo"))  # True
```

```python
# 8. count
# Cuenta el número de veces que una subcadena aparece en una cadena.

texto = "Hola Mundo Hola"
print(texto.count("Hola"))  # 2
```
